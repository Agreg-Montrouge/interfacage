```python
%matplotlib inline
import matplotlib.pyplot as plt

from tpmontrouge.instrument import Scope, get_all_connected_devices, get_first_device
```

## Connected to the scope
* If a single scope is connected use ``get_first_device``
* Otherwise if ``get_all_connected_devices``

You need to call the ``get_instrument`` method to accutally get the interface

```python
scope = get_first_device('Scope').get_instrument()
```

## Get data

If the scope is manually configured, c
* Check that the channel is active 
* get the channel and the use the ``get_waveform`` method
* the returnd object contains the data and can also perform a direct plot

```python
wfm = scope.channel[1].get_waveform()
```

```python
wfm.plot_matplotlib(fig=plt.figure())
```

```python
print(wfm.y_data.std())
```

## Change parameters of the scope

### Channel property

```python
print('    State :', scope.channel[1].state)
# On peut utiliser scope.channel[1].is_active()
print('    Impedance :', scope.channel[1].impedance)
print('    Coupling :', scope.channel[1].coupling)
print('    Scale :', scope.channel[1].scale)
```

```python
scope.channel[1].impedance = "Max" # OneMegOhm / 1E6 / FiftyOhm / 50 / Max / float('inf')
scope.channel[1].coupling = 'AC' # AC / DC / GND
scope.channel[1].scale = 0.001
```

### Horizontal property

```python
print('Scale', scope.horizontal.scale, 's/div')
print('Offset', scope.horizontal.offset, 's')
```

### Trigger

```python
print('Trigger source', scope.trigger.source)
print('Trigger level', scope.trigger.level)
```

```python
scope.trigger.slope = 'NegativeEdge' # or 'PositivEdge'
```
